```python
query = 'SELECT * FROM store_analysis'
```

```python
tradedata = pd.read_sql(query, conn)
```

```python
print(tradedata.head())
```

```python
StoreSize = tradedata['StoreSize']
TotalEmployees = tradedata['TotalEmployees']
YearlyRevenue = tradedata['YearlyRevenue']
Revenue = tradedata['Revenue']
scatter1 = plt.scatter(StoreSize, YearlyRevenue, color='darkblue', label = 'Yearly Revenue', alpha=0.3)
plt.xlabel("Store Size (Square Feet)") 
plt.ylabel("Yearly Revenue")
plt.twinx()
scatter2 = plt.scatter(StoreSize, TotalEmployees, color='green', label = 'Total Employees', alpha=0.3)
plt.ylabel("Total Employees")
plt.title("Relationship Between Store Size, Total Employees and Yearly Revenue")
plt.legend(handles=[scatter1, scatter2])
plt.show()
```
